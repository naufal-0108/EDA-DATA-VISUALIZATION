```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Meta Data Source = https://www.kaggle.com/datasets/aungpyaeap/supermarket-sales
Attribute information

- Invoice id: Computer generated sales slip invoice identification number.

- Branch: Branch of supercenter (3 branches are available identified by A, B and C).

- City: Location of supercenters.

- Customer type: Type of customers, recorded by Members for customers using member card and Normal for without member card.

- Gender: Gender type of customer.

- Product line: General item categorization groups - Electronic accessories, Fashion accessories, Food and beverages, Health and beauty, Home and lifestyle, Sports and travel.

- Unit price: Price of each product in ($).

- Quantity: Number of products purchased by customer.
 
- Tax: 5 % tax fee for customer buying.

- Total: Total price including tax.

- Date: Date of purchase (Record available from January 2019 to March 2019).

- Time: Purchase time (10am to 9pm).

- Payment: Payment used by customer for purchase (3 methods are available â€“ Cash, Credit card and Ewallet).

- COGS: Cost of goods sold.

- Gross margin percentage: Gross margin percentage.

- Gross income: Gross income.

- Rating: Customer stratification rating on their overall shopping experience (On a scale of 1 to 10).

```python
df = pd.read_csv('supermarket.csv')
print(df.info())
```

```python
display(df.head())
```

```python
display(df.tail())
```

```python
df['Date'] = df['Date'] + df['Time']
display(df.head())
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = "%m/%d/%Y%H:%M")
display(df.head())
```

```python
df_sorted = df.sort_values(by = 'Date')
```

```python
grouped1 = df_sorted.groupby('City').resample('d', on = 'Date')['gross income'].sum()
print(grouped1)
```

```python
sales_mandalay = grouped1.loc['Mandalay']
sales_Naypyitaw = grouped1.loc['Naypyitaw']
sales_Yangon = grouped1.loc['Yangon']
```

```python
print(sales_mandalay.values.mean())
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(3,1,figsize = (10,18))
ax[0].plot(sales_mandalay.index, sales_mandalay.values, label = 'Mandalay', color = 'red')
ax[1].plot(sales_Naypyitaw.index, sales_Naypyitaw.values, label = 'Naypyitaw', color = 'blue')
ax[2].plot(sales_Yangon.index, sales_Yangon.values, label = 'Yangon', color = 'green')

ax[0].axhline(sales_mandalay.values.mean(), label = 'Average Profits', linestyle = '--', color = 'black', alpha = 0.5)
ax[1].axhline(sales_Naypyitaw.values.mean(), label = 'Average Profits', linestyle = '--', color = 'black', alpha = 0.5)
ax[2].axhline(sales_Yangon.values.mean(), label = 'Average Profits', linestyle = '--', color = 'black', alpha = 0.5)

ax[0].set_title("Mandalay District's Profits" )
ax[1].set_title("Naypyitaw District's Profits")
ax[2].set_title("Yangon District's Profits")

ax[0].set_xticklabels([])
ax[1].set_xticklabels([])
ax[0].set_ylabel('Profit')
ax[1].set_ylabel('Profit')
ax[2].set_ylabel('Profit')

ax[0].legend(loc = 'upper left')
ax[1].legend()
ax[2].legend()
plt.xticks(rotation = 45)
plt.show()
```

```python
grouped2 = df_sorted.groupby('Product line').resample('d', on = 'Date')['gross income'].sum()
print(grouped2)
```

```python
profit_EA = grouped2.loc['Electronic accessories']
profit_FA = grouped2.loc['Fashion accessories']
profit_FB = grouped2.loc['Food and beverages']
profit_HB = grouped2.loc['Health and beauty']
profit_HL = grouped2.loc['Home and lifestyle']
profit_ST = grouped2.loc['Sports and travel']
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(6,1,figsize = (10,30))
ax[0].plot(profit_EA.index, profit_EA.values, label = 'Electronic Accessories', color = 'red')
ax[1].plot(profit_FA.index, profit_FA.values, label = 'Fashion Accessories', color = 'blue')
ax[2].plot(profit_FB.index, profit_FB.values, label = 'Food and Beverages', color = 'green')
ax[3].plot(profit_HB.index, profit_HB.values, label = 'Health and Beauty', color = 'purple')
ax[4].plot(profit_HL.index, profit_HL.values, label = 'Home and Lifestyle', color = 'cyan')
ax[5].plot(profit_ST.index, profit_ST.values, label = 'Sports and Travel', color = 'yellow')



ax[0].axhline(profit_EA.values.mean(), label = 'Average Profits', linestyle = '--', color = 'black', alpha = 0.5)
ax[1].axhline(profit_FA.values.mean(), label = 'Average Profits', linestyle = '--', color = 'black', alpha = 0.5)
ax[2].axhline(profit_FB.values.mean(), label = 'Average Profits', linestyle = '--', color = 'black', alpha = 0.5)
ax[3].axhline(profit_HB.values.mean(), label = 'Average Profits', linestyle = '--', color = 'black', alpha = 0.5)
ax[4].axhline(profit_HL.values.mean(), label = 'Average Profits', linestyle = '--', color = 'black', alpha = 0.5)
ax[5].axhline(profit_ST.values.mean(), label = 'Average Profits', linestyle = '--', color = 'black', alpha = 0.5)

ax[0].set_title("Electronic Accessories's Profits" )
ax[1].set_title("Fashion Accessories's Profits")
ax[2].set_title("Food and Beverages's Profits")
ax[3].set_title("Health and Beauty's Profits")
ax[4].set_title("Home and lifestyle's Profits")
ax[5].set_title("Sports and Travel's Profits")

ax[0].set_xticklabels([])
ax[1].set_xticklabels([])
ax[2].set_xticklabels([])
ax[3].set_xticklabels([])
ax[4].set_xticklabels([])

ax[0].set_ylabel('Profit')
ax[1].set_ylabel('Profit')
ax[2].set_ylabel('Profit')
ax[3].set_ylabel('Profit')
ax[4].set_ylabel('Profit')
ax[5].set_ylabel('Profit')

ax[0].legend()
ax[1].legend()
ax[2].legend()
ax[3].legend()
ax[4].legend()
ax[5].legend()
plt.xticks(rotation = 45)
plt.show()
```

```python
date =  ['2019-01-31','2019-02-28','2019-03-31']
month = ['January','February','March']

fig= plt.figure(figsize = (18,18))
for i in range(3):
    val = productline[productline['Date'].isin([date[i]])]
    ax = fig.add_subplot(2,2,i+1)
    ax.pie(val['gross income'], labels = val['Product line'],autopct = '%0.2f%%')
    ax.set_title(f'Propotion of Profit Based on Product Lines ( {month[i]} )', fontdict = {'fontsize' : 10})

plt.subplots_adjust(top = 0.8,bottom = 0.4, right = 0.7,wspace=0.9, 
                    hspace=0.2)
plt.show()
```
